## Design and Test a Model Architecture

### Pre-process the data

1. Convert to YUV color
2. Normalize data
3. Split into validation and training
4. Shuffle - Important

````python
import cv2

def convert_to_yuv(images):
    images_y = np.array([cv2.cvtColor(img, cv2.COLOR_BGR2YUV) for img in images])
    return images_y

def norm_images(images):
    return np.multiply(images, 1.0 / 255.0)

def preprocess_images(images):
    return norm_images(convert_to_yuv(images))

X_pre = preprocess_images(X)
X_test_pre = preprocess_images(X_test)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

X_train, X_validation, y_train, y_validation = train_test_split(X_pre, y, test_size=0.20, random_state=42)
X_train, y_train = shuffle(X_train, y_train)
````

### Conv Net

````python
from tensorflow.contrib.layers import flatten

def conv_layer(x, shape, strides, padding, mean, stddev):
    W = tf.Variable(tf.truncated_normal(shape, mean=mean, stddev=stddev))
    b = tf.Variable(tf.zeros(shape[-1]))
    conv = tf.nn.conv2d(x, W, strides=strides, padding=padding) + b
    return tf.nn.relu(conv)


def fully_connected(x, shape, mu, sigma, actiavation=True):
    W = tf.Variable(tf.truncated_normal(shape, mean=mu, stddev=sigma))
    b = tf.Variable(tf.zeros(shape[-1]))
    fc = tf.matmul(x, W) + b
    if actiavation:
        return tf.nn.relu(fc)
    else:
        return fc
    
def LeNet(x):    
    # Hyperparameters
    mu = 0
    sigma = 0.1
    padding = 'VALID'
    
    conv1 = conv_layer(x, (5,5,3,6), [1,1,1,1], padding, mu, sigma)
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding=padding)

    conv2 = conv_layer(conv1, (5,5,6,16), [1,1,1,1], padding, mu, sigma)
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')

    fc0   = flatten(conv2)
    fc1   = fully_connected(fc0, (400, 120), mu, sigma)
    fc2   = fully_connected(fc1, (120, 84), mu, sigma)
    return fully_connected(fc2, (84, 43), mu, sigma, False)
````

### Setup TensorFlow Variables

1. A placeholder is like a variable. Using None on an axis allows you to pass batches of any size.
2. One hot converts our labels to one hot vectors
3. Cross entropy with logits takes our logits from the network on does [cross entropy](https://en.wikipedia.org/wiki/Cross_entropy) on them.
4. Reduce mean just takes the average
5. Optimizer sets our gradient descent [optimizer](http://cs231n.github.io/neural-networks-3/) which we then minimize.
6. We then have a prediction (select highest probability) and an accuracy opteration in tensorflow.

````python
x_tf = tf.placeholder(tf.float32, (None, 32, 32, 3))
y_tf = tf.placeholder(tf.int32, (None))
dropout_keep_prob = tf.placeholder(tf.float32)
one_hot_y = tf.one_hot(y_tf, 43)

rate = 0.001

logits = LeNet(x_tf)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()
````

### Function to evaluate progress

1. Runs the accuracy_operation on the validation set and reports total accuracy.

````python
def evaluate(X_data, y_data, BATCH_SIZE):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x_tf: batch_x, y_tf: batch_y, dropout_keep_prob:1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

### Train our network

1. Initialize all variables in a tf session
2. Run our optimization for a certain number of EPOCHS using a BATCH_SIZE
3. Make sure to always shuffle each EPOCH data set
4. Run our training_operation from above with a feed_dict that fills in the values for our tf placeholders. Note: I pass DROP_PROB as an exaample even though I don't use dropout in this network. I do this because you have to make sure in validation that dropout probability is one (see the evaluate function).
5. Report validation accuracy and save model after each EPOCH

````python
EPOCHS = 5
BATCH_SIZE = 128
DROP_PROB = 0.5
SAVE_FILE = 'alex_net'

with tf.Session() as sess:
    sess.run(tf.initialize_all_variables())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x_tf: batch_x, y_tf: batch_y, dropout_keep_prob: DROP_PROB})
            
        validation_accuracy = evaluate(X_validation, y_validation, BATCH_SIZE)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
        saver.save(sess, SAVE_FILE)
        print("Model saved")
````
